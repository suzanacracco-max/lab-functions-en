## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
# your code here
# function
# accumulator starts in 1 because is a multiplication
# factorial - multiply in a loop 
# output

    total = 1
    i = 1
    
    while i <= n:
        total = total * i
        i = i + 1

    return total
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

```python
# This will test your function 
test_factorial(factorial)
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... ðŸ§`

```python
def mode_counter(arr):
#your code here
    max_count = 0
    mode_value = None

    for most_freq in arr:
        count = 0
        for x in arr:
            if x == most_freq:
                count += 1

        if count > max_count:
            max_count = count
            mode_value = most_freq

    return mode_value
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. ðŸ˜‰`

```python
#It measures how spread out the numbers in a list are.

#For a list of numbers:

    # - Compute the mean (average)
    # - Compute how far each number is from the mean
    # - Square those differences
    # - Average the squared differences (this is the variance)
    # - Take the square root of the variance
```

```python
def st_dev(list_sd):
#your code here
    # Step 1: calculate the mean
    total = 0
    count = 0

    for x in list_sd:
        total = total + x
        count = count + 1

    mean = total / count

    # Step 2: calculate sum of squared differences
    sq_diff_sum = 0

    for x in list_sd:
        diff = x - mean
        sq_diff_sum = sq_diff_sum + (diff * diff)

    # Step 3: variance
    variance = sq_diff_sum / (count - 1)

    # Step 4: standard deviation (square root)
    return variance ** 0.5
   
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
#your code here
    found=[False]*26
    for letter in string:
        if "A" <= letter <= "Z":
            letter = chr(ord(letter)+32) #converts a number into the character that number represents.
        if "a" <= letter <= "z":
            index = ord(letter) - ord ("a") # converts a character into its numeric (Unicode) value.
            found[index] = True
    for element in found:
        if not element:
            return False
    return True
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ðŸ¤ª`

```python
def sort_alpha(string):
#your code here
    words = []
    current_word = ""

    # Build list manually
    for ch in string:
        if ch == ',':
            words = words + [current_word]
            current_word = ""
        else:
            current_word = current_word + ch

    words = words + [current_word]
    # Manually count length
    n = 0
    for _ in words:
        n = n + 1
    # Bubble sort
    i = 0
    while i < n:
        j = 0
        while j < n - 1:
            if words[j] > words[j + 1]:
                temp = words[j]
                words[j] = words[j + 1]
                words[j + 1] = temp
            j = j + 1
        i = i + 1
    # Build output string manually
    result = ""
    first = True
    for w in words:
        if first:
            result = w
            first = False
        else:
            result = result + "," + w

    return result

```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
#your code here

    count = 0
    has_lower = False
    has_upper = False
    has_digit = False
    has_special = False

    for ch in password:
        count = count + 1

        # lowercase a-z
        if ch >= 'a' and ch <= 'z':
            has_lower = True

        # uppercase A-Z
        elif ch >= 'A' and ch <= 'Z':
            has_upper = True

        # digits 0-9
        elif ch >= '0' and ch <= '9':
            has_digit = True

        # anything else = special character
        else:
            has_special = True

    if count >= 8 and has_lower and has_upper and has_digit and has_special:
        return True
    else:
        return False
```

```python
# This will test your function 
test_pass(check_pass)
```
